```python
# Author: Krista Miller
# Title: Plastic Surgery Trends over time
# Inspiration: The Economist
# Original plot: https://www.economist.com/finance-and-economics/a-coup-in-guinea-adds-fuel-to-aluminiums-red-hot-rally/21804363
# Type: Line Chart

import datetime as dt
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import pandas as pd

prices = pd.read_excel('plastic_surg.xlsx', parse_dates=['Date'], index_col='Date')
prices.head()
```

```python
colors = ['#008bbc', '#3e647d', '#1a476f', '#EFF5F5']
colors_lines = ['#90353b', '#01a2d9', '#bfa19c', '#2E45B8', '#E2365B']

fig, ax = plt.subplots(figsize=(12, 12), frameon=True, facecolor=colors[3])

prices[['Breast Augmentation', 'Facelift', 'Rhinoplasty', 'Liposuction', 'Eyelid Surgery']].dropna().plot.line(figsize=(10, 10), ax=ax, color=colors_lines, linewidth=2, legend=True)

ax.spines[['top', 'right', 'left']].set_visible(False)

ax.set_facecolor(colors[3])

#the economist red square:
ax.plot([0.05, 0.075], [1, 1], color="#e3120b", linewidth=20, clip_on=False, transform=fig.transFigure)

ax.text(x=0.05, y=0.93, s="Plastic Surgery", fontsize=20, fontweight="bold", ha="left", transform=fig.transFigure)
ax.text(x=0.05, y=0.90, s="Top 5 plastic surgeries in US, 2006-2020", fontsize=16, ha="left", transform=fig.transFigure)

# xpos = dt.datetime.strptime('2016-01-01', '%Y-%m-%d')
# ypos = 325000
# ax.text(xpos, ypos, 'Breast Augmentation', fontsize=14, color='black', )

# xpos = dt.datetime.strptime('2016-01-01', '%Y-%m-%d')
# ypos = 100000
# ax.text(xpos, ypos, 'Face Lift', fontsize=14, fontweight='bold', color=colors_lines[0], ha='left')

# xpos = dt.datetime.strptime('2018-01-01', '%Y-%m-%d')
# ypos = 350000
# ax.text(xpos, ypos, 'Rhinoplasty', fontsize=14, color='black', )

#y-axis formatting
ax.yaxis.set_ticks_position('right')
ax.set_ylim([100000, 400000])
ax.set_yticks(range(100000, 400000, 50000))
ax.tick_params(axis="y", direction="out", length=0, color=colors[2], pad=5, labelsize=12)

#x-axis formatting
ax.set_xlabel('Year', fontdict={'fontsize': 16}, labelpad=7)
ax.xaxis.set_ticks_position('bottom')
ax.tick_params(axis='x', direction='out', length=12, pad=5, labelsize=12)
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_horizontalalignment('center')

#titles
ax.text(x=0.05, y=0.05, s="Source: American Society of Plastic Surgeons", fontsize=10, ha="left",
        transform=fig.transFigure)

ax.text(x=0.95, y=0.05, s="K.Miller", ha="right", transform=fig.transFigure,
        fontdict={'fontsize': 16, 'family': 'sans-serif', 'fontname': 'PT Serif Caption',
                  'color': 'black'})

ax.grid(axis='y')
plt.subplots_adjust(top=0.85, bottom=0.18, left=0.05, right=0.9)
plt.show()                                             
```
